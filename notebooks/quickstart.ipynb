## 2. Load the Search Engine

```python
import sys
sys.path.insert(0, '..')

from structural_isomorphism import StructuralSearch

# This loads the fine-tuned model and indexes the knowledge base
search = StructuralSearch()
print(search)
```

## 3. Query Examples

### Example 1: Feedback Loops

Describe a negative feedback system and find structural analogies:

```python
results = search.query(
    "恒温器检测到房间温度低于设定值就启动加热，温度达标后关闭，如此反复维持稳定",
    top_k=10
)

print("Query: Thermostat negative feedback loop")
print("=" * 60)
for i, r in enumerate(results, 1):
    print(f"{i:2d}. {r['name']:20s} ({r['domain']:10s}) | score: {r['score']:.3f}")
    print(f"    {r['description'][:80]}...")
    print()
```

### Example 2: S-Curve Growth

Describe a technology adoption pattern:

```python
results = search.query(
    "一种新产品刚上市时只有少数人尝鲜，然后口碑传播让用户快速增长，但当大多数潜在用户都已购买后增速放缓",
    top_k=10
)

print("Query: S-curve product adoption")
print("=" * 60)
for i, r in enumerate(results, 1):
    print(f"{i:2d}. {r['name']:20s} ({r['domain']:10s}) | score: {r['score']:.3f}")
    print(f"    {r['description'][:80]}...")
    print()
```

### Example 3: Collective Action Problem

Describe a social dilemma:

```python
results = search.query(
    "每个人都做出对自己最优的选择，但所有人的理性选择合在一起，反而导致了对所有人都不好的结果",
    top_k=10
)

print("Query: Tragedy of the commons / collective action failure")
print("=" * 60)
for i, r in enumerate(results, 1):
    print(f"{i:2d}. {r['name']:20s} ({r['domain']:10s}) | score: {r['score']:.3f}")
    print(f"    {r['description'][:80]}...")
    print()
```

## 4. Direct Embedding Access

You can also access the raw embeddings for custom analysis:

```python
from sentence_transformers import util

# Encode two descriptions and compare
texts = [
    "电路中电压越高，通过导线的电流就越大，两者之间保持着固定的换算关系",
    "水管中水压越大，水流就越急，压力和流量之间有着稳定的对应",
    "蜜蜂在花丛中采蜜后回到蜂巢，通过舞蹈告诉同伴食物的方向和距离",
]

embeddings = search.encode(texts)
print(f"Embedding shape: {embeddings.shape}")

# Pairwise similarities
for i in range(len(texts)):
    for j in range(i + 1, len(texts)):
        sim = float(util.cos_sim(embeddings[i], embeddings[j]))
        print(f"Similarity [{i}] vs [{j}]: {sim:.4f}")
        
print("\n[0] Ohm's law (electricity)")
print("[1] Fluid flow (hydraulics) -- same structure as [0]")
print("[2] Bee dance (biology) -- different structure")
```

## 5. Cross-Domain Discovery

Find potentially unknown structural connections in the knowledge base:

```python
pairs = search.find_cross_domain_pairs(threshold=0.75, max_pairs=10)

print(f"Top {len(pairs)} cross-domain structural connections:")
print("=" * 80)
for i, p in enumerate(pairs, 1):
    a, b = p['item_a'], p['item_b']
    print(f"{i:2d}. [{a['domain']}] {a['name']}")
    print(f"    <-> [{b['domain']}] {b['name']}")
    print(f"    Similarity: {p['similarity']:.3f}")
    print()
```
